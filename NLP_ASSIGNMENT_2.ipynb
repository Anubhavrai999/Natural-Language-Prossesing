# 1. Create a Doc object from the file owlcreek.txt


```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
with open('/content/owlcreek.txt','r',encoding='utf-8') as f:
  x=f.read()
doc=nlp(x)
```

# 2. How many tokens are contained in the file?

```python
print(f"Number of tokens: {len(doc)}")
```

#3. How many sentences are contained in the file?


```python
print(f"Number of sentences: {len(list(doc.sents))}")
```

#4. Print the second sentence in the document


```python
s1 = list(doc.sents)
s2 = s1[1]
print("\nSecond sentence:")
print(s2.text)
```

#5. For each token in the sentence above, print its text, POS tag, dep tag and lemma.


```python
print(f"{'Text tag':<15} | {'POS tag':<10} | {'Dep tag':<15} | {'Lemma tag':<15}")
for token in s2:
    print(f"{token.text:<15} {token.pos_:<10} {token.dep_:<15} {token.lemma_:<15}")
```

#6. Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text.


```python
from spacy.matcher import Matcher
```

```python
matcher = Matcher(nlp.vocab)
pattern = [{"LOWER": "swimming"}, {"LOWER": "vigorously"}]
matcher.add("Swimming", [pattern])
matches = matcher(doc)
print(f"\nNumber of 'swimming vigorously' matches: {len(matches)}")
```

#7. Print the text surrounding each found match.


```python
for match_id, start, end in matches:
    span = doc[(start - 5):(end + 5)]
    print(span , "\n" , "-"*10)
```
